1. Read and Display an Image

   Use cv2.imread() to load an image and cv2.imshow() to display it.


```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('catimg.jpg')
```

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

2. Draw a Circle on an Image

   Use cv2.circle() to draw a circle at a specified location on the image.

```python
    center_coordinates = (100, 100)  
    radius = 50                    
    color = (255, 0, 0)           
    thickness = 5                 

    cv2.circle(img1, center_coordinates, radius, color, thickness)
    plt.imshow(img1)
    plt.show()
```

3. Draw a Line on an Image

   Use cv2.line() to draw a straight line between two points on the image

```python

    start_point = (50, 50)  
    end_point = (200, 50)   
    line_color = (0, 255, 0)  
    line_thickness = 3   
    
    img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    cv2.line(img2, start_point, end_point, line_color, line_thickness)
    
    plt.imshow(img2)
    plt.show()

```

4. Draw a Rectangle on an Image

   Use cv2.rectangle() to draw a rectangle with specified corners.

```python

    top_left = (60, 80)    
    bottom_right = (180, 150) 
    rectangle_color = (255, 0, 0) 
    rectangle_thickness = 5        
    
    img3 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    cv2.rectangle(img3, top_left, bottom_right, rectangle_color, rectangle_thickness)
    plt.imshow(img3)
    plt.show()
```

5. Draw Multiple Shapes on an Image

   Combine different shapes like circles, lines, and rectangles on the same image.

```python
img4 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.line(img4, (50, 50), (200, 50), (255, 0, 0), 5)
cv2.rectangle(img4, (60, 60), (160, 160), (0, 255, 0), 3)
cv2.circle(img4, (120, 120), 40, (0, 0, 255), -1)
plt.show()
```

```python

#  Display Image Sections
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 

height, width, _ = image.shape

top_left = image[:height // 2, :width // 2]
plt.figure(figsize=(6, 6))
plt.title('Top-Left Section')
plt.imshow(top_left)
plt.axis('off')  
plt.show()

top_right = image[:height // 2, width // 2:]
plt.figure(figsize=(6, 6))
plt.title('Top-Right Section')
plt.imshow(top_right)
plt.axis('off')
plt.show()

```

```python
# Display the original image and Blured image
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
height, width, _ = image.shape
plt.imshow(image)

# Gaussian Blur
blurred = cv2.GaussianBlur(image, (15, 15), 0)
plt.figure()
plt.title('Gaussian Blurred Image')
plt.imshow(blurred)
plt.show()
```

```python

#  Convert Image to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)  
plt.figure()
plt.title('Grayscale Image')
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python

#  Resize the Image
new_width, new_height = width // 2, height // 2
resized = cv2.resize(image, (new_width, new_height))
plt.figure()
plt.title('Resized Image')
plt.imshow(resized)
plt.show()
```

```python

# Rotate the Image
center = (width // 2, height // 2)
angle = 45  
scale = 1.0 
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated = cv2.warpAffine(image, rotation_matrix, (width, height))
plt.figure()
plt.title('Rotated Image')
plt.imshow(rotated)
plt.show()
```

```python

#  Save the Modified Image
cv2.imwrite('modified_image.jpg', cv2.cvtColor(rotated, cv2.COLOR_RGB2BGR))

```

```python

#  Display Image in Multiple Color Channels
b, g, r = cv2.split(cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
plt.figure()
plt.title('Blue Channel')
plt.imshow(b, cmap='gray')

plt.figure()
plt.title('Green Channel')
plt.imshow(g, cmap='gray')

plt.figure()
plt.title('Red Channel')
plt.imshow(r, cmap='gray')
plt.show()
```

```python

# Draw Text on the Image
text_image = image.copy()
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(text_image, 'Hello!', (50, 50), font, 1, (0, 255, 0), 2, cv2.LINE_AA)
plt.figure()
plt.title('Image with Text')
plt.imshow(text_image)

plt.show()
```
